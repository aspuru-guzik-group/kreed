```python
import os
os.chdir('../..')

from src.baseline import values, bin_edges

import numpy as np
import matplotlib.pyplot as plt
```

```python
bin_mids = (bin_edges[1:] + bin_edges[:-1]) / 2
```

```python
# increase font size
plt.rcParams.update({'font.size': 14})

# Create a 1x2 grid of subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True, dpi=300)  # Adjust figsize as needed

# Plot data in the first subplot
axes[0].plot(bin_mids, values['qm9'])
axes[0].set_xlim([0, 7.5])
axes[0].set_title('QM9')
axes[0].set_ylabel('Probability density')
axes[0].set_xlabel("Heavy atom distance ($\AA$)")

# Plot data in the second subplot
axes[1].plot(bin_mids, values['geom'])
axes[1].set_xlim([0, 14])
axes[1].set_title('GEOM')
axes[1].set_xlabel("Heavy atom distance ($\AA$)")

# Add spacing between subplots
plt.tight_layout()
```
